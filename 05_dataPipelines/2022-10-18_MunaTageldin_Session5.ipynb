```python
import pandas as pd 
import numpy as np 
data=pd.read_csv("threads_vs_buffer_size.csv", index_col=False)
print(data)
```

## The impact of varying parallel threads on mean image rate 


```python
#print(data["mean_image  "])
data.groupby(["parallel_threads"])["mean_image  "].plot(x ='parallel threads',style='-o',title="Mean Image Rate",kind='line')
```

## The impact of varying buffer size on mean image rate 

```python
data.groupby(["buffer_size"])["mean_image  "].plot(x ='buffer size',style='-o',title="Mean Image Rate",kind='line')
```
