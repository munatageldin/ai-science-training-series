## import data

```python
import pandas as pd 
import numpy as np 
data_1=pd.read_csv("data_1.csv")
data_2=pd.read_csv("data_2.csv")
data_4=pd.read_csv("data_4.csv")
data_8=pd.read_csv("data_8.csv")
print(data_1)
#data.columns =['Name', 'Code', 'Age', 'Weight','tail']
```

## Plots for 1 GPU

```python
from matplotlib import pyplot as plt
fig, axes = plt.subplots(figsize=(10,10), nrows=2, ncols=1)
data_1['train_acc'].plot(legend=False,kind="line",style="-o",x="epoch",title="Train_accuracy",color="blue",ax=axes[0])
data_1['valid_acc'].plot(legend=False,kind="line",style="-o",title="Validation_accuracy",color="blue",ax=axes[1])
plt.savefig("accuracy_1.pdf")
  
plt.show()
```
